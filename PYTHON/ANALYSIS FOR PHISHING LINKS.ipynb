```python
import re

def extract_urls(text):
    return re.findall(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', str(text))


df['Comment'] = df['Comment'].apply(lambda x: ' '.join(extract_urls(x)))


print(df)
```

```python
df.head(n=20)
```

```python

df = df[df['Comment'].str.strip().str.strip('"') != '']

df.reset_index(drop=True, inplace=True)

print(df)
```

```python
df.to_csv("C:/Users/durga/Desktop/SPU/DS 600/PROJECT/FINAL/PHISHING_LINKS.CSV", index=False)
```

```python
# Function to check if a URL is a phishing link
def is_phishing(url):
    # Regular expression to match common phishing patterns
    pattern = r'(https?://(?:www\.)?[\w\-\.]+\.\w+/?)'
    if re.match(pattern, url):
        return True
    return False


df['Phishing'] = df['Comment'].apply(lambda x: '🚩' if is_phishing(x) else '')


print(df)
```
